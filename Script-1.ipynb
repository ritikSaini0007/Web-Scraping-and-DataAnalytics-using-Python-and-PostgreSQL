```python
#For Proper Scraping

import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://books.toscrape.com/catalogue/page-{}.html"

data = []

rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

for page in range(1, 51):  # 50 pages total
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")

    for book in books:

        # Title
        title = book.h3.a["title"]

        # Price
        price = book.find("p", class_="price_color").text

        # Rating
        rating_class = book.p["class"][1]
        rating = rating_map.get(rating_class, 0)

        # Availability
        availability = book.find("p", class_="instock availability").text.strip()

        data.append([title, price, rating, availability])

df = pd.DataFrame(data, columns=["Title", "Price", "Rating", "Availability"])

print("Total Books Scraped:", len(df))
print(df.head())

df.to_csv("all_books.csv", index=False)
```
