```python
from sqlalchemy import create_engine
import pandas as pd

df = pd.read_csv("cleaned_books.csv")

engine = create_engine(
    "postgresql://postgres:1234@localhost:5432/bookstore"
)

df.to_sql(
    "books22",
    engine,
    if_exists="append",
    index=False
)

print("Data Imported Successfully!")
```
