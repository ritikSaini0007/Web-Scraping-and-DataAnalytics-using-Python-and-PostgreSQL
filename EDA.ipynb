```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("cleaned_books.csv")
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
print(df.describe())
```

```python
print(f"Total Books: {len(df)}")
```

```python
print(f"Average Price: £{df['Price'].mean():.2f}")
```

```python
expensive = df.loc[df["Price"].idxmax()]

print(expensive)
```

```python
cheap = df.loc[df["Price"].idxmin()]

print(cheap)
```

```python
rating_counts = df["Rating"].value_counts().sort_index()

plt.figure(figsize=(8,5))

plt.bar(rating_counts.index.astype(str),
        rating_counts.values)

plt.title("Books by Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Books")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Price"], bins=15)

plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Frequency")

plt.show()
```

```python
avg_price = df.groupby("Rating")["Price"].mean()

plt.figure(figsize=(8,5))

plt.bar(avg_price.index.astype(str),
        avg_price.values)

plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (£)")

plt.show()
```

```python
top10 = df.nlargest(10, "Price")

plt.figure(figsize=(10,6))

plt.barh(top10["Title"], top10["Price"])

plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")

plt.gca().invert_yaxis()

plt.show()
```

```python
cheap10 = df.nsmallest(10, "Price")

plt.figure(figsize=(10,6))

plt.barh(cheap10["Title"], cheap10["Price"])

plt.title("Top 10 Cheapest Books")
plt.xlabel("Price (£)")

plt.gca().invert_yaxis()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df["Price"])

plt.title("Book Price Boxplot")
plt.ylabel("Price (£)")

plt.show()
```

```python
availability = df["Availability"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    availability.values,
    labels=availability.index,
    autopct="%1.1f%%"
)

plt.title("Book Availability")

plt.show()
```
