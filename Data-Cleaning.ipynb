```python
import pandas as pd

df = pd.read_csv("clean_books.csv")

print("=" * 50)
print("Dataset Loaded Successfully")
print("=" * 50)
```

```python
# Create a Copy

clean_df = df.copy()
```

```python

print("\nDataset Shape:")
print(clean_df.shape)

print("\nColumn Names:")
print(clean_df.columns.tolist())

print("\nData Types:")
print(clean_df.dtypes)
```

```python
#missing values 
print("\nMissing Values:")
print(clean_df.isnull().sum())
```

```python
# Check Duplicate Records
duplicates = clean_df.duplicated().sum()

print(f"\nDuplicate Records: {duplicates}")

if duplicates > 0:
    clean_df.drop_duplicates(inplace=True)
    print("Duplicates Removed Successfully.")
else:
    print("No Duplicate Records Found.")
```

```python
# Clean Price Column
# Remove currency symbol
clean_df["Price"] = (
    clean_df["Price"]
    .str.replace("Â£", "", regex=False)
    .str.strip()
)

# Convert Price to float
clean_df["Price"] = clean_df["Price"].astype(float)

print("\nPrice Column Cleaned Successfully.")

```

```python
# Clean Availability Column
clean_df["Availability"] = clean_df["Availability"].str.strip()
print("\nAvailability Column Standardized.")

# Verify Data Types
print("\nUpdated Data Types:")
print(clean_df.dtypes)

#Summary Statistics
print("\nSummary Statistics:")
print(clean_df.describe())

# Check Unique Values
print("\nUnique Ratings:")
print(sorted(clean_df["Rating"].unique()))

print("\nAvailability Status:")
print(clean_df["Availability"].unique())
```

```python
#Final Dataset Information
print("\nFinal Dataset Shape:")
print(clean_df.shape)

print("\nFirst Five Rows:")
print(clean_df.head())


# Save Cleaned Dataset
clean_df.to_csv("cleaned_books.csv", index=False)
print("\nCleaned Dataset Saved Successfully!")
print("Location: data/cleaned_books.csv")
```
